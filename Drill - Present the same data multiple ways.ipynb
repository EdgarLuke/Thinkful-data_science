```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('plane_crashes_data.csv')
```

```python
df.head()
```

```python
sns.set_style('whitegrid')
sns.distplot(df.fatalities, bins=20, kde=False)
plt.title('Distribution of Fatalities by Counts')
plt.show()
```

```python
sns.distplot(df.fatalities, hist=False, kde=True, rug=True)
plt.title('Kernel Distribution Estimate with Rug Plot')
plt.show()
```

```python
sns.violinplot(x='fatalities', data=df)

plt.show()
```

```python
sns.set()

g = sns.factorplot("fatalities", data=df, aspect=3, kind="count")
g.set_xticklabels(step=20)

plt.show()
```

```python
df.head()
```

```python
sns.set_style('whitegrid')

sns.jointplot(x='aboard', y='fatalities', size=6, data=df, space=0.2)

plt.show()
```

```python
g = sns.stripplot(x='fatalities', y='aboard', data=df)
g.set_xticklabels(step=20, labels='')

plt.show()
```

```python
sns.jointplot(x='fatalities', y='aboard', data=df, kind='scatter', color='red')
plt.show()
```

```python
df.head()
```

```python
sns.boxplot(x=df['month'], y=df['fatalities'], data=df)

plt.show()
```

```python
sns.pairplot(df, hue='month', palette='Blues')
plt.show()
```

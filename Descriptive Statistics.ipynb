```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame()
df['ints'] = np.random.randint(1, 25, 100)
chisquare = np.random.chisquare(df['ints'], 100)

plt.hist(chisquare)
# Add a vertical line at the mean
plt.axvline(chisquare.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at 1 standard deviation ABOVE the mean
plt.axvline(chisquare.mean() + chisquare.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at 1 standard deviation BELOW the mean
plt.axvline(chisquare.mean() - chisquare.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
exponential = np.random.exponential(4, 100)

plt.hist(exponential, color='orange')
# Add a vertical line at the mean
plt.axvline(exponential.mean(), color='r', linestyle='solid', linewidth=2)

# Plot the median to check better fit
plt.axvline(np.median(exponential), color='b', linestyle='solid', linewidth=2)
# Add a vertical line at 1 standard deviation ABOVE the mean
plt.axvline(exponential.mean() + exponential.std(), color='r', linestyle='dashed', linewidth=2)

# Add a vertical line at 1 standard deviation BELOW the mean
plt.axvline(exponential.mean() - exponential.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
weibull = np.random.weibull(5, 100)

plt.hist(weibull, color='c')
# Add a vertical line at the mean
plt.axvline(weibull.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at 1 standard deviation ABOVE the mean
plt.axvline(weibull.mean() + weibull.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at 1 standard deviation BELOW the mean
plt.axvline(weibull.mean() - weibull.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
x = np.random.normal(5, 0.5, 100)
y = np.random.normal(10, 1, 100)

z = x + y
plt.hist(z, color='black')
# Add a vertical line at the mean
plt.axvline(z.mean(), color='g', linestyle='solid', linewidth=2)

# Add a vertical line at 1 standard deviation ABOVE the mean
plt.axvline(z.mean() + z.std(), color='g', linestyle='dashed', linewidth=2)

# Add a vertical line at 1 standard deviation BELOW the mean
plt.axvline(z.mean() - z.std(), color='g', linestyle='dashed', linewidth=2)
plt.show()
```
